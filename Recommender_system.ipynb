```python
movie_df= pd.read_csv('/content/sample_data/movies.csv',encoding = "ISO-8859-1")
rating_df= pd.read_csv('/content/sample_data/ratings.csv',encoding = "ISO-8859-1")
feature_df= pd.read_csv('/content/sample_data/tags.csv',encoding = "ISO-8859-1")
```

```python
movie_df['genres_list']= movie_df['genres'].str.split('|')
movie_df1 = movie_df.explode('genres_list')
movie_df1 =pd.get_dummies(movie_df1,columns=['genres_list'],prefix='',prefix_sep='')
movie_df2 =movie_df1.groupby(['movieId', 'title', 'genres']).sum()
movie_df2.reset_index(inplace=True)
movie_df2.head(2)
```

```python
rating_pivot = rating_df.pivot_table(index=['userId'],columns=['movieId'],values=['rating'],fill_value=0)
rating_pivot.columns =rating_pivot.columns.droplevel()
rating_pivot_norm = rating_pivot.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)
rating_pivot_norm.head(2)
```
